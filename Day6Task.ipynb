### Task 2: List Problems 
1. Find maximum and minimum  
2. Find second largest  
3. Remove duplicates  
4. Count even and odd numbers 

```python
# Find maximum and minimum
nums = [10, 20, 5, 40]

largest=nums[0]
smallest=nums[0]
for i in nums:
    if i>largest:
        largest=i
print("Largest Number in list:",largest)
for i in nums:
    if i<smallest:
        smallest=i
print("Smallest Number in list:",smallest)
```

```python
# Simple logic
print(max(nums))
print(min(nums))
```

```python
# Find Second largets
nums = [10, 20, 5, 40] 
nums.sort()
print(nums[-2])
```

```python
# Remove duplicates
nums = [1, 2, 2, 3, 4, 4] 
 
unique = list(set(nums)) 
print(unique)
```

```python
# Count even and odd numbers
numbers = [1, 2, 3, 4, 5, 6]
even =[n for n in numbers if n%2==0]
odd=[n for n in numbers if n%2!=0]
print("Even Numbers:",even)
print("Odd Numbers:",odd)
```

### Task 3: String Problems 
1. Reverse string  
2. Check palindrome  
3. Count characters  
4. Count vowels and consonants  

```python

text = input("Enter your String: ")

while True:
    print("\n1. Reverse string")  
    print("2. Check palindrome")  
    print("3. Count characters")  
    print("4. Count vowels and consonants") 
    print("5. Exit")
    
    choice = int(input("Enter your choice: "))

    if choice == 1:
        print("Reverse Of a String:", text[::-1])
        
    elif choice == 2:
        if text[::-1] == text:
            print("String is palindrome")
        else:
            print("String is not a palindrome")

    elif choice == 3:
        count = {}
        for ch in text:
            count[ch] = count.get(ch, 0) + 1
        print("Character count:", count)

    elif choice == 4:
        vowels = 0
        consonants = 0

        for ch in text:
            if ch in "aeiouAEIOU":
                vowels += 1
            elif ch.isalpha():
                consonants += 1

        print("Vowels:", vowels)
        print("Consonants:", consonants)

    elif choice == 5:
        print("Exiting program...")
        break

    else:
        print("Please Select Proper Choice")
```

### Task 4: Dictionary Problems 
1. Create student list  
2. Find topper  
3. Calculate average  
4. Filter passed students

```python
students = {}

n = int(input("Enter number of students: "))

for i in range(n):
    name = input("Enter student name: ")
    marks = int(input("Enter marks: "))
    students[name] = marks

print("Student List:", students)
```

```python
# Find Topper
topper = max(students, key=students.get)
print("Topper:",topper,":", students[topper])
```

```python
#Calculate average
avg = sum(students.values()) / len(students)
print("Average Marks:", avg)
```

```python
# Filter average Students
above_avg = {name: marks for name, marks in students.items() if marks >= avg}
print("Students above average:", above_avg)
```

```python
below_avg =  {name: marks for name, marks in students.items() if marks< avg}
print("Students have below average:",below_avg)
```
