```python
"""
  Example Problem 
     Find sum of numbers from 1 to N 
Step 1: Logic 
• Take input N  
• Loop from 1 to N  
• Add numbers 
"""

#Code

numbers=int(input("Enter the number:\n"))
total=0

for i in range(1,numbers+1):
 total+=i

print(total)
         
                  
```

```python
"""
Working with Lists (Advanced Logic) 
  Find Maximum Number 
  """
nums = [10, 20, 5, 40]

largest=nums[0]
for i in nums:
    if i>largest:
        largest=i
print(largest)
```

```python
#Optimize the code 
print(max(nums))
```

```python
 # Find Second Largest Number 
nums = [10, 20, 5, 40] 
nums.sort()
print(nums[-2])
```

```python
#Count Characters (how many times each character appears)
text = "hello" 
 
count = {} 
 
for ch in text: 
    count[ch] = count.get(ch, 0) + 1 
 
print(count)
```

```python
 # Check Palindrome 
text = input("Enter string: ") 
 
if text == text[::-1]: 
    print("Palindrome") 
else: 
    print("Not Palindrome")
```

```python
#Part 4: Working with Dictionaries (Advanced) 
#  Find Top Student 
students = [ 
    {"name": "A", "marks": 70}, 
    {"name": "B", "marks": 90} 
] 
 
top = students[0] 
 
for s in students: 
    if s["marks"] > top["marks"]: 
        top = s 
 
print(top)
```

```python
  #Group Data 
marks = [50, 80, 90, 40] 
 
passed = [m for m in marks if m >= 50] 
print(passed)
```
